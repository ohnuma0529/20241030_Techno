```python
class VideoProcessor:
    def __init__(self, video_path):
        self.video_path = video_path
        self.cap = cv2.VideoCapture(video_path)
        self.frame_count = 0

    def release(self):
        self.cap.release()

    def process_frames(self, process_function):
        while True:
            ret, frame = self.cap.read()
            if not ret:
                break
            process_function(frame, self.frame_count)
            self.frame_count += 1
        self.release()
        print("全てのフレームを処理しました。")


class PoseEstimator:
    def __init__(self, model_path, output_folder):
        self.model = YOLO(model_path)
        self.output_folder = output_folder
        os.makedirs(self.output_folder, exist_ok=True)

    def estimate_pose(self, frame, frame_count):
        results = self.model(frame)
        annotated_frame = results[0].plot()
        frame_filename = os.path.join(self.output_folder, f"frame_{frame_count:04d}.jpg")
        cv2.imwrite(frame_filename, annotated_frame)


class BBoxEstimator:
    def __init__(self, model_path, output_folder):
        self.model = YOLO(model_path)
        self.output_folder = output_folder
        os.makedirs(self.output_folder, exist_ok=True)

    def estimate_bbox(self, frame, frame_count):
        results = self.model(frame)
        for result in results:
            if result.names[result.boxes.cls[0].item()] == "person":
                annotated_frame = result.plot()
                frame_filename = os.path.join(self.output_folder, f"frame_{frame_count:04d}.jpg")
                cv2.imwrite(frame_filename, annotated_frame)


class DepthEstimator:
    def __init__(self, encoder, output_folder):
        self.model_configs = {
            'vits': {'encoder': 'vits', 'features': 64, 'out_channels': [48, 96, 192, 384]},
            'vitb': {'encoder': 'vitb', 'features': 128, 'out_channels': [96, 192, 384, 768]},
            'vitl': {'encoder': 'vitl', 'features': 256, 'out_channels': [256, 512, 1024, 1024]},
            'vitg': {'encoder': 'vitg', 'features': 384, 'out_channels': [1536, 1536, 1536, 1536]}
        }
        self.encoder = encoder
        self.model = DepthAnythingV2(**self.model_configs[encoder])
        self.model.load_state_dict(torch.load(f"Depth-Anything-V2/depth_anything_v2_{encoder}.pth", map_location="cpu"))
        self.model.eval()
        self.model = self.model.to('cuda')
        self.output_folder = output_folder
        os.makedirs(self.output_folder, exist_ok=True)

    def estimate_depth(self, frame, frame_count):
        depth = self.model.infer_image(frame)
        depth_min = depth.min()
        depth_max = depth.max()
        depth_normalized = (depth - depth_min) / (depth_max - depth_min) * 255
        depth_normalized = depth_normalized.astype(np.uint8)
        frame_filename = os.path.join(self.output_folder, f"frame_{frame_count:04d}.jpg")
        cv2.imwrite(frame_filename, depth_normalized)


def main():
    video_path = "./input/6387-191695740.mp4"
    
    # 各推定器の初期化
    pose_estimator = PoseEstimator("weight/yolov8n-pose.pt", "output/keypoint")
    bbox_estimator = BBoxEstimator("weight/yolov8n.pt", "output/bbox")
    depth_estimator = DepthEstimator("vits", "output/depth")

    # 動画処理の開始
    video_processor = VideoProcessor(video_path)
    
    print("Pose推定中...")
    video_processor.process_frames(pose_estimator.estimate_pose)
    
    print("BBOX推定中...")
    video_processor.process_frames(bbox_estimator.estimate_bbox)

    print("Depth推定中...")
    video_processor.process_frames(depth_estimator.estimate_depth)


if __name__ == "__main__":
    main()
```

```python
class VideoProcessor:
    def __init__(self, video_path):
        self.video_path = video_path
        self.cap = cv2.VideoCapture(video_path)
        self.frame_count = 0

    def release(self):
        self.cap.release()

    def process_frames(self, process_function):
        results = []
        while True:
            ret, frame = self.cap.read()
            if not ret:
                break
            result = process_function(frame)
            results.append(result)
            self.frame_count += 1
        self.release()
        return results


class PoseEstimator:
    def __init__(self, model_path):
        self.model = YOLO(model_path)

    def estimate_pose(self, frame):
        results = self.model(frame)
        annotated_frame = results[0].plot()
        return annotated_frame


class BBoxEstimator:
    def __init__(self, model_path):
        self.model = YOLO(model_path)

    def estimate_bbox(self, frame):
        results = self.model(frame)
        for result in results:
            if result.names[result.boxes.cls[0].item()] == "person":
                annotated_frame = result.plot()
                return annotated_frame
        return frame  # 人が検出されなかった場合は元のフレームを返す


class DepthEstimator:
    def __init__(self, encoder):
        self.model_configs = {
            'vits': {'encoder': 'vits', 'features': 64, 'out_channels': [48, 96, 192, 384]},
            'vitb': {'encoder': 'vitb', 'features': 128, 'out_channels': [96, 192, 384, 768]},
            'vitl': {'encoder': 'vitl', 'features': 256, 'out_channels': [256, 512, 1024, 1024]},
            'vitg': {'encoder': 'vitg', 'features': 384, 'out_channels': [1536, 1536, 1536, 1536]}
        }
        self.encoder = encoder
        self.model = DepthAnythingV2(**self.model_configs[encoder])
        self.model.load_state_dict(torch.load(f"Depth-Anything-V2/depth_anything_v2_{encoder}.pth", map_location="cpu"))
        self.model.eval()
        self.model = self.model.to('cuda')

    def estimate_depth(self, frame):
        depth = self.model.infer_image(frame)
        depth_min = depth.min()
        depth_max = depth.max()
        depth_normalized = (depth - depth_min) / (depth_max - depth_min) * 255
        depth_normalized = depth_normalized.astype(np.uint8)

        # DEPTH結果を3チャンネルに変換
        depth_colored = cv2.cvtColor(depth_normalized, cv2.COLOR_GRAY2BGR)
        return depth_colored


def main():
    video_path = "./input/6387-191695740.mp4"
    
    # 各推定器の初期化
    pose_estimator = PoseEstimator("weight/yolov8n-pose.pt")
    bbox_estimator = BBoxEstimator("weight/yolov8n.pt")
    depth_estimator = DepthEstimator("vits")

    # 動画処理の開始
    video_processor = VideoProcessor(video_path)

    all_results = []
    
    print("推定中...")
    cnt = 0
    for frame in video_processor.process_frames(lambda f: f):
        # 各推定を実行
        bbox_result = bbox_estimator.estimate_bbox(frame)
        pose_result = pose_estimator.estimate_pose(frame)
        depth_result = depth_estimator.estimate_depth(frame)

        # 結合画像の生成
        top_row = np.hstack((frame, bbox_result))
        bottom_row = np.hstack((pose_result, depth_result))
        combined_image = np.vstack((top_row, bottom_row))

        # 結合画像を保存
        output_folder = "output/combined"
        os.makedirs(output_folder, exist_ok=True)
        frame_filename = os.path.join(output_folder, f"combined_frame_{cnt:04d}.jpg")
        cv2.imwrite(frame_filename, combined_image)
        cnt += 1
    print("全てのフレームを処理しました。")


if __name__ == "__main__":
    main()
```
